Question:<br>
Figure out which reports are safe.The final result is how many reports are safe?<br>
Each row contains a report, the columns have a value resembling a level.<br>
<br>
Safe criteria:<br>
<ul><li>Levels are all increasing or all decreasing<br>
<li>Adjacent levels may differ min 1 and max 3 steps<br>
</ul><br>
Steps:<br>
<ol><li>Load input and store each line as a sub element of a list<br>
<li>Iterate over each sublist, checking if the second value is equal, higher or lower<br>
<li>When the second value is equal, label it unsafe<br>
<li>When the second value is higher or lower, determine the distance between them, when x < 1 or x > 3, label it unsafe<br>
<li>When the second value is higher or lower, and 1 <= x <= 3 is true, go to the next value and take the same steps<br>
<li>When all values are checkt and not labelled as unsafe, the report is labelled as safe, increment a counter with 1.<br>
<li>When all input is checked, return the total value of the counter<br>




Result part 1:

Run 2: 314 v
